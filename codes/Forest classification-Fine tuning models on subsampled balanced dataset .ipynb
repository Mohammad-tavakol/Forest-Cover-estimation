```python
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import Normalizer

scaler=StandardScaler()
X1=scaler.fit_transform(X)
plt.boxplot(X1);
```

```python
from sklearn.model_selection import train_test_split
xtrain , xtest, ytrain, ytest=train_test_split(X1,Y,test_size=0.3,random_state=42)
#data shape
print(xtrain.shape, xtest.shape, ytrain.shape, ytest.shape)
```

```python
#models
import seaborn as sns
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
```

```python
from sklearn import metrics
from sklearn.metrics import classification_report,confusion_matrix, ConfusionMatrixDisplay
dtc = DecisionTreeClassifier(splitter='best',max_depth=100,criterion='gini',
                           min_samples_split=3, min_samples_leaf=2, random_state=42)
DTC = dtc.fit(xtrain, ytrain)
DTC_trscore=DTC.score(xtrain,ytrain)
ypred_dt=DTC.predict(xtest)
print("Decision Tree Train Accuracy:",DTC_trscore)
print("Decision Tree Test Accuracy:", metrics.accuracy_score(ytest, ypred_dt))
print('----------------------------------------')
print('* Decision Tree Classification Report')
print(classification_report(ytest, ypred_dt))
print('----------------------------------------')
print('* Decision Tree Confusion Matrix')
# Plot confusion matrix
from sklearn.metrics import ConfusionMatrixDisplay, confusion_matrix
# Plot confusion matrix
sns.set(font_scale = 1.1)
cm = confusion_matrix(ytest, ypred_dt)
cmn = cm.astype('float') / cm.sum(axis=1)[:, np.newaxis]
sns.heatmap(cmn, annot=True,cmap='RdPu',  fmt='.2f',xticklabels=["C1", "C2", "C3", "C4","C5","C6","C7"], yticklabels=["C1", "C2", "C3", "C4","C5","C6","C7"])

```
